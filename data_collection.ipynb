```python
import pandas as pd

#An empty list to hold individual dataframes from each file
data_list = []

#Columns that are extracted from each file
columns = ['route',
 'name',
 'type',
 'departure_time',
 'arrival_time',
 'duration',
 'price',
 'seats_available',
 'rating']

#List of CSV files to be read
files_list = [
            'JKSRTC_buses.csv',
            'KAAC_buses.csv',
            'West_bengal_buses.csv',
            'NORTH_BENGAL_buses.csv',
            'Bihar_buses.csv',
            'Chandigarh_buses.csv',
            'Assam_buses.csv',
            'Punjab_buses.csv',
            'south_Bengal_buses.csv',
            'Telangana_buses.csv',
            'Kerala_buses.csv',
            'Goa_buses.csv',
            'Andhra_Pradesh_buses.csv']

# Using loop through each file in the files_list
for file in files_list:
    df = pd.read_csv(file)
    df = df[columns]
    data_list.append(df)

#  Concatenate all dataframes in the list into a single dataframe  
file_data = pd.concat(data_list)
file_data = file_data.reset_index(drop = True)
file_data[['from_place', 'to_place']] = file_data['route'].str.split(' to ', expand=True)
```

```python
df = pd.DataFrame(file_data)

# Save the DataFrame to a CSV file
df.to_csv("All_Bus_data_Collated.csv", index=False)

print("Data has been saved to All_Bus_data_Collated.csv")
```

```python
data=pd.read_csv("All_Bus_data_Collated.csv")
data
```

```python
#function to remove alphabetic characters from a string column

def change_name(x):
    return x.str.replace(r'[A-Za-z]+', '', regex=True)
data['price']=data['price'].transform(change_name)
```

```python
#Applying the 'change_name' function to the 'seats_available' column of the DataFrame
data['seats_available']=data['seats_available'].transform(change_name)
data
```

```python
#Converting the 'rating' column to numeric values, coercing errors to NaN
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating']=data['rating'].fillna(0)
data['rating']=data['rating'].astype('float64')
data
```

```python
#Converting the 'seats_available' column to numeric values, coercing errors to NaN
data['seats_available'] = pd.to_numeric(data['seats_available'], errors='coerce')
data['seats_available']=data['seats_available'].fillna(0)
data['seats_available']=data['seats_available'].astype('int64')
data
```

```python
#Converting the 'price' column to numeric values, coercing errors to NaN
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['price']=data['price'].fillna(0)
data['price']=data['price'].astype('int64')
data
```
